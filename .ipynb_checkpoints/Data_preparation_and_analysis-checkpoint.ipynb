###  √âtape 4:  Classification des cotes 

**c‚Äôest quoi exactement ?**

Une **cote** est un code unique attribu√© √† chaque document d'une biblioth√®que servant de :  
- *Adresse physique* : Indique l'emplacement exact sur les √©tag√®res  
- *Cl√© de recherche* : Permet de retrouver le document dans le catalogue  
- *Identifiant th√©matique* : Classe les documents par domaine de connaissance

**Exemple de granularit√©**

1)- Une cote comme <code> 330.114/01 </code> signifie :

<code>330</code> = √©conomie

<code>.114</code> = un sous-th√®me

<code>/01</code> = une num√©rotation locale ou exemplaire

2)- 

 <code> "330.114 </code>" ‚Üí 
   -    33 0 : √âconomie
  -     .1 : Th√©orie √©conomique
 -      .14 : March√© du travail

##### Syst√®me Dewey (Notre Choix)  pour la classification des cotes
*Classification D√©cimale Dewey* - Syst√®me universellement adopt√© dans les biblioth√®ques publiques :

| Code  | Th√©matique                | Exemples Courants                          |
|-------|---------------------------|--------------------------------------------|
| `000` | **G√©n√©ralit√©s**           | Informatique, Encyclop√©dies, Bibliographie |
| `100` | **Philosophie**           | √âthique, Psychologie, Logique              |
| `300` | **Sciences Sociales**     | √âconomie, Droit, √âducation, Politique      |
| `500` | **Sciences Exactes**      | Math√©matiques, Physique, Chimie, Astronomie |
| `800` | **Litt√©rature**           | Romans, Po√©sie, Th√©√¢tre, Critique litt√©raire |



````python
# Dictionnaire des classes principales Dewey
dewey_classes = {
    '000': 'G√©n√©ralit√©s',
    '100': 'Philosophie et psychologie',
    '200': 'Religion',
    '300': 'Sciences sociales',
    '400': 'Langues',
    '500': 'Sciences naturelles et math√©matiques',
    '600': 'Technologie (sciences appliqu√©es)',
    '700': 'Arts et loisirs',
    '800': 'Litt√©rature',
    '900': 'Histoire et g√©ographie'
}

# Fonction pour extraire la classe Dewey √† partir de la cote
def get_dewey_class(cote):
    if pd.isna(cote):
        return 'Inconnue'
    # Extraction de la partie num√©rique de la cote (ex: "330.114/01" ‚Üí "330")
    try:
        # Prend les 3 premiers chiffres avant le point ou le slash
        code = str(cote).split('/')[0].split('.')[0][:3]
        code_int = int(code)
        # Ram√®ne au multiple de 100 (ex: 330 ‚Üí 300)
        base = str((code_int // 100) * 100).zfill(3)
        return dewey_classes.get(base, 'Hors Dewey')
    except:
        return 'Inconnue'

# Appliquer aux deux datasets
df_fr['Classe Dewey'] = df_fr['Cote'].apply(get_dewey_class)
df_ar['Classe Dewey'] = df_ar['Cote'].apply(get_dewey_class)

# Afficher un aper√ßu des r√©sultats
print(" Fran√ßais :")
print(df_fr[['Cote', 'Classe Dewey']].head())

print("\n Arabe :")
print(df_ar[['Cote', 'Classe Dewey']].head())
````

###  √âtape 5: Visualisation des donn√©es- Distribution des documents par classe Dewey

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Appliquer un style propre
sns.set(style="whitegrid")

# Fonction de visualisation
def plot_dewey_distribution(df, title):
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(data=df, y='Classe Dewey', order=df['Classe Dewey'].value_counts().index, palette='viridis')
    plt.title(f'Distribution des documents par classe Dewey ‚Äì {title}')
    plt.xlabel("Nombre de documents")
    plt.ylabel("Classe Dewey")
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='edge', fontsize=9)
    plt.tight_layout()
    plt.show()

# Visualisation pour df_fr
plot_dewey_distribution(df_fr, 'Collection en Fran√ßais')

# Visualisation pour df_ar
plot_dewey_distribution(df_ar, 'Collection en Arabe ')
````

###  √âtape 6: Reformatage de l'inventaire

````python
# Fonction pour reformater l'inventaire
def reformatter_inventaire(col):
    return col.astype(str).apply(lambda x: x if x.startswith('FSR_') else f'FSR_{x}')

# Application sur df_fr et df_ar
df_fr['Inventaire'] = reformatter_inventaire(df_fr['Inventaire'])
df_ar['Inventaire'] = reformatter_inventaire(df_ar['Inventaire'])
````
