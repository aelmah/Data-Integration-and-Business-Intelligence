###  Ã‰tape 4:  Classification des cotes 

**câ€™est quoi exactement ?**

Une **cote** est un code unique attribuÃ© Ã  chaque document d'une bibliothÃ¨que servant de :  
- *Adresse physique* : Indique l'emplacement exact sur les Ã©tagÃ¨res  
- *ClÃ© de recherche* : Permet de retrouver le document dans le catalogue  
- *Identifiant thÃ©matique* : Classe les documents par domaine de connaissance

**Exemple de granularitÃ©**

1)- Une cote comme <code> 330.114/01 </code> signifie :

<code>330</code> = Ã©conomie

<code>.114</code> = un sous-thÃ¨me

<code>/01</code> = une numÃ©rotation locale ou exemplaire

2)- 

 <code> "330.114 </code>" â†’ 
   -    33 0 : Ã‰conomie
  -     .1 : ThÃ©orie Ã©conomique
 -      .14 : MarchÃ© du travail

##### SystÃ¨me Dewey (Notre Choix)  pour la classification des cotes
*Classification DÃ©cimale Dewey* - SystÃ¨me universellement adoptÃ© dans les bibliothÃ¨ques publiques :

| Code  | ThÃ©matique                | Exemples Courants                          |
|-------|---------------------------|--------------------------------------------|
| `000` | **GÃ©nÃ©ralitÃ©s**           | Informatique, EncyclopÃ©dies, Bibliographie |
| `100` | **Philosophie**           | Ã‰thique, Psychologie, Logique              |
| `300` | **Sciences Sociales**     | Ã‰conomie, Droit, Ã‰ducation, Politique      |
| `500` | **Sciences Exactes**      | MathÃ©matiques, Physique, Chimie, Astronomie |
| `800` | **LittÃ©rature**           | Romans, PoÃ©sie, ThÃ©Ã¢tre, Critique littÃ©raire |



````python
# Dictionnaire des classes principales Dewey
dewey_classes = {
    '000': 'GÃ©nÃ©ralitÃ©s',
    '100': 'Philosophie et psychologie',
    '200': 'Religion',
    '300': 'Sciences sociales',
    '400': 'Langues',
    '500': 'Sciences naturelles et mathÃ©matiques',
    '600': 'Technologie (sciences appliquÃ©es)',
    '700': 'Arts et loisirs',
    '800': 'LittÃ©rature',
    '900': 'Histoire et gÃ©ographie'
}

# Fonction pour extraire la classe Dewey Ã  partir de la cote
def get_dewey_class(cote):
    if pd.isna(cote):
        return 'Inconnue'
    # Extraction de la partie numÃ©rique de la cote (ex: "330.114/01" â†’ "330")
    try:
        # Prend les 3 premiers chiffres avant le point ou le slash
        code = str(cote).split('/')[0].split('.')[0][:3]
        code_int = int(code)
        # RamÃ¨ne au multiple de 100 (ex: 330 â†’ 300)
        base = str((code_int // 100) * 100).zfill(3)
        return dewey_classes.get(base, 'Hors Dewey')
    except:
        return 'Inconnue'

# Appliquer aux deux datasets
df_fr['Classe Dewey'] = df_fr['Cote'].apply(get_dewey_class)
df_ar['Classe Dewey'] = df_ar['Cote'].apply(get_dewey_class)

# Afficher un aperÃ§u des rÃ©sultats
print(" FranÃ§ais :")
print(df_fr[['Cote', 'Classe Dewey']].head())

print("\n Arabe :")
print(df_ar[['Cote', 'Classe Dewey']].head())
````

###  Ã‰tape 5: Visualisation des donnÃ©es- Distribution des documents par classe Dewey

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Appliquer un style propre
sns.set(style="whitegrid")

# Fonction de visualisation
def plot_dewey_distribution(df, title):
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(data=df, y='Classe Dewey', order=df['Classe Dewey'].value_counts().index, palette='viridis')
    plt.title(f'Distribution des documents par classe Dewey â€“ {title}')
    plt.xlabel("Nombre de documents")
    plt.ylabel("Classe Dewey")
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='edge', fontsize=9)
    plt.tight_layout()
    plt.show()

# Visualisation pour df_fr
plot_dewey_distribution(df_fr, 'Collection en FranÃ§ais')

# Visualisation pour df_ar
plot_dewey_distribution(df_ar, 'Collection en Arabe ')
````

###  Ã‰tape 6: Reformatage de l'inventaire

````python
# Fonction pour reformater l'inventaire
def reformatter_inventaire(col):
    return col.astype(str).apply(lambda x: x if x.startswith('FSR_') else f'FSR_{x}')

# Application sur df_fr et df_ar
df_fr['Inventaire'] = reformatter_inventaire(df_fr['Inventaire'])
df_ar['Inventaire'] = reformatter_inventaire(df_ar['Inventaire'])
````
